### 2. Count of products without prices and with prices in each Product Type, Category, Level 1 

#### Count of products without prices and with prices in each Product Type

```python
query_str='''
      Select product_type, count(product_type) total,sum(case when price_string is null then 1 else 0 end) without_price,
      sum(case when price_string is not null then 1 else 0 end) with_price
      from df
      group by product_type
      '''
duckdb.query(query_str).to_df()
```

#### Count of products without prices and with prices in each Category

```python
query_str='''
      Select category, count(category) total,sum(case when price_string is null then 1 else 0 end) without_price,
      sum(case when price_string is not null then 1 else 0 end) with_price
      from df
      group by category
      '''
duckdb.query(query_str).to_df()
```

#### Count of products without prices and with prices in each Level 1

```python
query_str='''
      Select level_1, count(level_1) total,sum(case when price_string is null then 1 else 0 end) without_price,
      sum(case when price_string is not null then 1 else 0 end) with_price
      from df
      group by level_1
      '''
duckdb.query(query_str).to_df()
```

### 3. Correct Product Prices in the correct format (eg: $56) wherever possible and separate them into currency and value columns.

```python
df['price_string']=df["price_string"].apply(lambda x: x if (~(str(x).find('$'))) else (x if pd.isna(x) else '$'+str(x)))
df_dollar=df["price_string"].str.split("", n = 2, expand = True)
df_dollar.columns=['first','sec','third']
df_dollar.drop(columns='first',axis=1,inplace=True)
df['currency']=df_dollar['sec']
df['price_value']=df_dollar['third']
df.head()
```

### 4. List out the categories with average price of product.

```python
df['price_value']=df['price_value'].fillna(0).astype('float')
df_cat=df.groupby('category').agg({'price_value':'mean'}).reset_index().rename(columns={'price_value':'avg_price_per_category'})

df_cat
```
